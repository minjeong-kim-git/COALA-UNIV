```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# (1, 90) (2, 100) (3, 50)
plt.plot([1,2,3], [90,100, 50])
plt.show()
```

```python
plt.plot(['Math', 'Korean', 'English'], [90, 100, 50])

plt.plot(['Math', 'Korean', 'English'], [70, 60, 90])

plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm')

plt.legend(['Choi', 'Kim'])

plt.show()
```

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')

plt.bar(x, y, width = 0.6, color = 'red')
```

```python
plt.plot([1, 2, 3], [90, 100, 50], marker = '+', markersize = 30, linestyle = '--')

plt.grid(True)
plt.axis([1, 3, 80, 100])

plt.show()
```
