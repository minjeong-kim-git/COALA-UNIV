```python
import pandas as pd

df = pd.read_csv('data/house.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
house_data = df.drop(['id', 'date'], axis = 1)
house_data.head()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (20, 15))
sns.heatmap(house_data.corr(), annot = True, fmt = '.2f', square = True)
plt.show()
```

```python
house_data.corr()
```

```python
sns.catplot(data = house_data, x = 'bathrooms', y = 'price', aspect = 2)
```

```python
sqft_max = house_data['sqft_living'].max()
```

```python
print('거실 면적 최댓값:', sqft_max)
```

```python
house_data['price'].max()
```

```python
print('집 값의 최솟값:', house_data['price'].min())
print('집 값의 최댓값:', house_data['price'].max())
print('집 값의 평균값:', house_data['price'].mean())
```
